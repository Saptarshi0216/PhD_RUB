```python
import matplotlib.pyplot as plt
import numpy as np

###########################################

#lifetimes in e-13 s, the smallest pie has its radius set to 0.5 (& area=0.25), with the areas of pie-charts of particles proportional to the lifetimes
# Here we calculate what the radii of the pies must be in accordance to their lifetimes

Xi0= 1.519                  #lifetime of charm hadron Xi0_c which decays the quickest in units of e-13s
C= (0.25/Xi0)               #the particle with the shortest life-time has been given an area of 0.25, thereby a radius of 0.5. We have ignored the Pi
                            # as it is just a scaling factor for our purposes
#hadrons
Xi_0= (1.519*C)**0.5
Lamda_cp = (2.015*C)**0.5
Xi_pp_cc= (2.56*C)**0.5
Omega0_c = (2.68*C)**0.5
Xi_cp= (4.53*C)**0.5
#mesons
D_pm= (10.33*C)**0.5
D0= (4.103*C)**0.5
D_pms= (5.04*C)**0.5
B_pm= (5.1*C)**0.5

###########################################
#particle specific plots
#Color coding: Up:orange, Down: crimson, Strange: violet, Charm: Green, Bottom: Cyan

#Lambda+= udc
labels = 'Up', 'Down', 'Charm'
colors=['orange','crimson', 'green']
sizes = [33.33, 33.33, 33.33]
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig, ax = plt.subplots()
ax.pie(sizes, radius= Lamda_cp, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
#ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title(label='$\lambda_c^+$ baryon',loc='left',fontsize=16)
plt.legend(shadow=True)

#Xi++_cc= dcc
fig1, ax1 = plt.subplots()
labels =  'Down', 'Charm'
colors= ['crimson','green']
sizes = [33.33, 66.67]
ax1.pie(sizes, radius= Xi_pp_cc, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$\Xi_{cc}^{++}$ baryon',loc='left',fontsize=16)
plt.legend(shadow=True)


#Xi_0= dsc
fig2, ax2 = plt.subplots()
labels =  'Down', 'Strange', 'Charm'
colors= ['crimson','violet','green']
sizes = [33.33, 33.33,33.33]
ax2.pie(sizes, radius= Xi_0, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$\Xi_{c}^0$ baryon',loc='left',fontsize=16)
plt.legend(shadow=True)


#Omega0_c= ssc
fig3, ax3 = plt.subplots()
labels =  'Strange', 'Charm'
colors= ['violet','green']
sizes = [66.67,33.33]
ax3.pie(sizes, radius= Omega0_c, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$\Omega_{c}^0$ baryon',loc='left',fontsize=16)
plt.legend(shadow=True)


#Xi_cp= usc
fig4, ax4 = plt.subplots()
labels =  'Up','Strange', 'Charm'
colors= ['orange','violet','green']
sizes = [33.33,33.33,33.33]
ax4.pie(sizes, radius= Xi_cp, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$\Xi_{c}^{+}$ baryon',loc='left',fontsize=16)
plt.legend(shadow=True)


#D_pm= cd
fig5, ax5 = plt.subplots()
labels =  'Charm/anti-Charm','anti-Down/Down'
colors= ['green','crimson']
sizes = [50,50]
ax5.pie(sizes, radius= D_pm, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$D^{\pm}$ meson',loc='left',fontsize=16)
plt.legend(loc=(1,1),shadow=True)


#D0= cu
fig6, ax6 = plt.subplots()
labels =  'Charm/anti-Charm','anti-Up/Up'
colors= ['green','orange']
sizes = [50,50]
ax6.pie(sizes, radius= D0, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$D^0$ meson',loc='left',fontsize=16)
plt.legend(loc=(0.8,0.8),shadow=True)


#D_pms= cs
fig7, ax7 = plt.subplots()
labels =  'Charm/anti-Charm','anti-Strange/Strange'
colors= ['green','violet']
sizes = [50,50]
ax7.pie(sizes, radius= D_pms, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$D^{\pm}_s$ meson',loc='left',fontsize=16)
plt.legend(loc=(0.8,0.8),shadow=True)


#B_pm= cb
fig8, ax8 = plt.subplots()
labels =  'Charm/anti-Charm','anti-Bottom/Bottom'
colors= ['green','cyan']
sizes = [50,50]
ax8.pie(sizes, radius= B_pm, labels=labels, autopct='%1.1f%%', startangle=90, colors= colors,wedgeprops = {"linewidth": 1, "edgecolor": "white"},shadow=True)
plt.title(label='$B^{\pm}$ meson',loc='left',fontsize=16)
plt.legend(loc=(0.8,0.8),shadow=True)


plt.tight_layout()
plt.show()

###########################################
```
